* Data **CerahHujan**, **Penguapan**, **KecSemburanAngin**, **KecAngin**, bersifat **Skewed** karena memiliki nilai lebih dari kurang lebih 0.5

# Random Forest Regression

```python
#Pandas series to choose target and features
X = datan.iloc[:, :-1]
y = datan.iloc[:, -1] #y here is the data in the last column which is the Temperature
X.shape
```

```python
y.shape
```

**Importing model and Splitting data**

```python
#Splitting Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)
#Importing model
from sklearn.ensemble import RandomForestRegressor
```

**The Model**

```python
#Naming the model and fitting the data
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

**Accuracy Matrix**

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
#RSME
np.sqrt(mean_squared_error(y_test, y_pred))
```
